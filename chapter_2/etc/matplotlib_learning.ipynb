Let's generate some sample data for a line plot:

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2*x)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.set_title("An example graph")
ax.set_xlabel("X data")
ax.set_ylabel("Y data")
ax.set(
    xlim=(0,8),
    ylim=(0,8),
    xticks=np.arange(1,8),
    yticks=np.arange(1,8),
)
ax.plot(x, y, linewidth=2.0)
```

```python
ages = np.arange(20,51, 10)
children_born = np.array([5,2,6,0])
print(f"{ages=} {children_born=}")
```

```python
fig, ax = plt.subplots()
ax.plot(ages, children_born)
ax.set_title("Avg children born per age")
ax.set_xlabel("Age")
ax.set_ylabel("Births")
```

```python
years_of_experience = np.arange(1,11)
avg_salary = np.linspace(75000, 500000, 10)
print(f"{years_of_experience=} {avg_salary=}")
```

```python
fig, ax = plt.subplots()
ax.bar(years_of_experience, avg_salary)
ax.set_title("Avg salary per years of experience")
ax.set_xlabel("Years of Experience")
ax.set_ylabel("Average Salary")
```
