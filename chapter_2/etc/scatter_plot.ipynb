```python
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
diabetes = load_diabetes()
description = diabetes["DESCR"]
print(description)
```

```python
import pandas as pd

diabetes_dataframe = pd.DataFrame(diabetes["data"][:,:4], columns=diabetes.feature_names[:4])
pd.plotting.scatter_matrix(diabetes_dataframe, c=diabetes["target"], marker="o", figsize=(20,20))
```
