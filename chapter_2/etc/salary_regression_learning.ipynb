```python
import pandas as pd

salary_df = pd.read_csv("Salary_Data.csv")
salary_df.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
X = salary_df.drop("Salary", axis=1).values
y = salary_df["Salary"].values

def format_func(value, tick_number):
    # This function converts large numbers to thousands (K)
    # and millions (M) for better readability
    if value >= 1_000_000:
        return f'{value//1_000_000}M'
    elif value >= 1000:
        return f'{value//1000}K'
    else:
        return str(value)

fig, ax = plt.subplots()
ax.set(
    title="Average Salary per Years of Experience",
    xlabel="Years of Experience",
    ylabel="Average Salary"
)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(format_func))
ax.plot(X, y, marker="o")
```

```python
# k-NN Regression
from sklearn.neighbors import KNeighborsRegressor



knr = KNeighborsRegressor(n_neighbors=1)
knr.fit(X, y)
```

```python
knr.predict([[3.75]])
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
```

```python
lr.predict([[3.75]])
```

```python
# lets take a look at the linear regression slop compared to the actual data points
import numpy as np

m = lr.coef_
b = lr.intercept_

x2 = np.linspace(0, 10, 100)
y2 = m * x2 + b

fig, ax = plt.subplots()
ax.grid(True)
ax.plot(x2,y2, color="red")
ax.scatter(X, y, color="green")

```
