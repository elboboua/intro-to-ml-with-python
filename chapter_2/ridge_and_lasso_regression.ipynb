```python
from sklearn.model_selection import train_test_split
from mglearn.datasets import load_extended_boston
from sklearn.linear_model import Ridge

X, y = load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

Let's take a look at the data with a scatter matrix

```python
import pandas as pd
boston_df = pd.DataFrame(X, columns=[f"feature_{i}" for i in range(X.shape[1])])
pd.plotting.scatter_matrix(boston_df.iloc[:,:5], c=y, marker="o", figsize=(20,20))
```

```python
rdg = Ridge()
rdg.fit(X_train, y_train)
train_score = rdg.score(X_train, y_train)
test_score = rdg.score(X_test, y_test)
print(f"Training score: {train_score:.2f}\tTesting score: {test_score:.2f}")
```

This is a little too good on the training score, let's increase alpha

```python
rdg = Ridge(alpha=5.0)
rdg.fit(X_train, y_train)
train_score = rdg.score(X_train, y_train)
test_score = rdg.score(X_test, y_test)
print(f"Training score: {train_score:.2f}\tTesting score: {test_score:.2f}")
```

This is a little too poor on testing, let's go towards the middle

```python
rdg = Ridge(alpha=0.1)
rdg.fit(X_train, y_train)
train_score = rdg.score(X_train, y_train)
test_score = rdg.score(X_test, y_test)
print(f"Training score: {train_score:.2f}\tTesting score: {test_score:.2f}")
```

Let's try with lasso

```python
from sklearn.linear_model import Lasso
lso = Lasso(alpha=0.005, max_iter=1000000)
lso.fit(X_train, y_train)
train_score = lso.score(X_train, y_train)
test_score = lso.score(X_test, y_test)
print(f"Training score: {train_score:.2f}\tTesting score: {test_score:.2f}")
```
